##  7.1 Writing a function piecewiseLinTransform

```python
def piecewiseLinTransform(img,k1,k2,a,b):
    img = img.astype(np.float64)/255
    i = img.shape[0]
    j = img.shape[1]
    new = np.zeros((i,j,3), dtype = float) # the input image we are taking 3 channel image
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            for k in range(3):
                real = img[i][j][k]
                temp = float(k1)*(img[i][j][k]) + float(k2)
                if(real > a and real < b):
                    new[i][j][k] = temp
                else:
                    new[i][j][k] = real
                new[i][j][k] = new[i][j][k] * 255
    new = new.astype(np.uint8)
    return new
```
